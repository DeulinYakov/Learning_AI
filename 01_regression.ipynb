```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math

%matplotlib inline
```

```python
plt.style.use('default')
```

```python
#Значенияh
lr = 0.001
epochs = 3000
W = 0.4
X_train = [6.3, 8, 9.8, 14.2, 16, 17.3, 23]
Y_train = [40, 60, 90.1, 137, 117, 130.9, 154.1]

X_test = [12.7, 20]
Y_test = [115, 146]

first_W = W
```

```python
i = 0
while i < epochs:
  for x, y in zip(X_train, Y_train):
    y_new = W * x
    E = y - y_new
    W = W + lr *(E / x)
  i += 1
print(W)
```

```python
def pred(W, X):
  return [ x * W for x in X] 
```

```python
fig, ax = plt.subplots(figsize=(15,10), dpi=72)
x_lim = (0,25)
ax.set_title('linear', pad='15', fontsize=24)
ax.set_xlabel('X', fontsize=16)
ax.set_ylabel('Y', fontsize=16)

ax.set(xlim=(0,25), ylim=(0,160))

ax.tick_params(axis='both', labelsize=14)

#Тикеры Х
ax.xaxis.set_minor_locator(ticker.MultipleLocator(2))
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))

#Тикеры У
ax.yaxis.set_minor_locator(ticker.MultipleLocator(10))
ax.yaxis.set_major_locator(ticker.MultipleLocator(5))

# Настройка сетки
ax.grid(True, color='0.85', linestyle='-')
ax.grid(which='minor', color='1', linestyle='-')
ax.grid(which='major', color='0.95', linestyle='-')

# Диаграмма рассеяния
ax.scatter([6.3, 8, 9.8, 12.7, 14.2, 16, 17.3, 20, 23], 
           [40, 60, 90.1, 115, 137, 117, 130.9, 146, 154],
           color='tomato', 
           s=50, zorder=2,
           label='Точки')

ax.plot(x_lim, pred(first_W, x_lim))
ax.plot(x_lim, pred(W, x_lim))
print(pred(W, X_test))
print(Y_test)
#plt.show()
plt.savefig('my_plot.png', transparent=True)
```
