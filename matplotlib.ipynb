```python
# Сначала необходимо подключить библиотеку matplotlib
import matplotlib.pyplot as plt

# Библиотека для настройки шкал
import matplotlib.ticker as ticker

# Для Jupyter - отображение в ноутбуке
%matplotlib inline
```

```python
# Чтобы не устанавливать каждый раз настройки
# для графиков, можно использовать или готовые
# стили, или созданные вами
# plt.style.use('ggplot')
# plt.style.use('default')
```

```python
# Matplotlib строит графики 
# внутри Figure - окнах или виджетах Jupyter

# Каждый Figure может содержать одну или несколько
# координатных плоскостей (Axes). 
```

```python
# Чтобы создать Figure с координатными плоскостями
# используется ф-ия pyplot.subplots

# Чтобы отобразить данные используется функция 
# Axes.plot()

# Создаем окно, содержащее одну координатную плоскость
fig, ax = plt.subplots(figsize=(15,10), dpi=72)
# figsize - размер графика в дюймах
# dpi - разрешение изображения (точек на дюйм)
# 72 - стандартное экранное разрешение
# 300 - разрешение для печати

# Название графика
ax.set_title('Линейный график', pad='15', fontsize=24)
# pad - смещение названия от верха графика

# Подписи осей
ax.set_xlabel('X', fontsize=16)
ax.set_ylabel('Y', fontsize=16)

# Диапазоны осей
ax.set(xlim=(0,5), ylim=(0,5))


# Настройка шкал -----------------------------------

# Размер шрифта
ax.tick_params(axis='both', labelsize=14)

# По оси X. 
# Основные значения
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
# Промежуточные значения
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.5))

# По оси Y
# Основные значения
ax.yaxis.set_major_locator(ticker.MultipleLocator(2))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(1))
# -------------------------------------------------

# Настройка сетки
ax.grid(True, color='0.85', linestyle='-')
ax.grid(which='minor', color='0.95', linestyle='-')
# про цвета можно почитать тут:
# https://matplotlib.org/stable/tutorials/colors/colors.html


# Формируем графики ------------------------------

# Обычный график
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], 
        color='yellowgreen', linewidth=2, 
        label='График')

# Диаграмма рассеяния
ax.scatter([0.5, 1.5, 2.5, 3.5], [1.7, 3, 2, 3.8],
           color='tomato', 
           s=50, zorder=2,
           label='Точки')

# Аннотация
ax.annotate('local max', 
            xy=(2, 4), 
            xytext=(1, 3.8),
            fontsize=14,
            arrowprops=dict(arrowstyle='->, head_width=0.5, head_length=1', 
                            color='black',
                            connectionstyle='arc3, rad=-0.5',
                            shrinkA=5, shrinkB=5,
                            linewidth=1)
            )

# Текст
ax.text(3.5, 1, 
        'Произвольный\nтекст', 
        {'fontsize':14, 'color':'grey'})
# ---------------------------------------------


# Легенда
# Метод legend берет подписи графиков из атрибута label метода plot,
# поэтому его вызов должен быть после описания всех графиков
# (после всех методов plot)
ax.legend(loc='best', fontsize=14)


# Отображаем график
plt.show()

# Также, график можно сохранить
# plt.savefig('my_plot.png', transparent=True)



# Если нужна специфичная координатная плоскость,
# то можно разнести их создание
# fig = plt.figure(figsize=(15,10), dpi=72)
# ax = Axes3D(fig)
```
